```python
import matplotlib.pyplot as plt
import numpy as np
# Veri
x = np.linspace(0, 10, 100)
y = np.sin(x)
# Ã‡izgi Grafik
plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o')
plt.title('Sine Fonksiyonu')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
kategoriler = ['A', 'B', 'C', 'D']
degerler = [10, 15, 7, 20]
# Ã‡ubuk Grafik
plt.bar(kategoriler, degerler, color=['red', 'green', 'blue', 'purple'])
plt.title('Kategori BazÄ±nda DeÄŸerler')
plt.xlabel('Kategoriler')
plt.ylabel('DeÄŸerler')
plt.show()
```

```python
import matplotlib.pyplot as plt
veri = np.random.randn(1000)
# Histogram
plt.hist(veri, bins=30, color='orange', edgecolor='black')
plt.title('Veri DaÄŸÄ±lÄ±mÄ±')
plt.xlabel('DeÄŸerler')
plt.ylabel('Frekans')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 25]
# Daire Grafik
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['cyan', 'magenta', 'yellow', 'green'])
plt.title('Kategori DaÄŸÄ±lÄ±mÄ±')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
# Ä°lk Grafik
axs[0].plot(x, y1, 'r-', label='Sin(x)')
axs[0].set_title('Sine Fonksiyonu')
axs[0].set_xlabel('X Ekseni')
axs[0].set_ylabel('Y Ekseni')
axs[0].grid(True)
axs[0].legend()
# Ä°kinci Grafik
axs[1].plot(x, y2, 'b-', label='Cos(x)')
axs[1].set_title('Cosine Fonksiyonu')
axs[1].set_xlabel('X Ekseni')
axs[1].set_ylabel('Y Ekseni')
axs[1].grid(True)
axs[1].legend()
plt.tight_layout()
plt.show()
```
